```python
import pydicom
import os 
import glob
import shutil
```

```python
dicomfolder = "/Users/kayti/Desktop/pycom_sort_test/dcm/" 
niioutput = "/Users/kayti/Desktop/pycom_sort_test/nii/"
```

```python
subjids = os.listdir(dicomfolder)
if '.DS_Store' in subjids:
  subjids.remove('.DS_Store')
```

```python
for ids in subjids:  
  path = dicomfolder + ids + "/**/*"
  dcmspath = glob.glob(path, recursive=True)
  dcmspathfile = dcmspath[100]
  dcmpath = os.path.dirname(dcmspathfile)
  dcms = os.listdir(dcmpath)
  if '.DS_Store' in dcms:
      dcms.remove('.DS_Store')
  for dcm in dcms:
      dcmhdr = pydicom.dcmread(dcmpath + '/' + dcm)
      pn = str(dcmhdr.ProtocolName)
      sn = str(dcmhdr.SeriesNumber)
      if os.path.exists(dicomfolder + '/' + ids + '/' + sn + '_' + pn):
          oldfile = os.path.join(dcmpath + '/' + dcm)
          newfile = os.path.join(dicomfolder + '/' + ids + '/' + sn +  '_' + pn + '/' + dcm)
          os.rename(oldfile,newfile)
      else: 
          os.mkdir(dicomfolder + '/' + ids + '/' + sn +  '_' + pn)
          oldfile = os.path.join(dcmpath + '/' + dcm)
          newfile = os.path.join(dicomfolder + '/' + ids + '/' + sn +  '_' + pn + '/' + dcm)
          os.rename(oldfile,newfile)
      for entry in os.scandir(dicomfolder + '/' + ids):
        if os.path.isdir(entry.path) and not os.listdir(entry.path) :
          os.rmdir(entry.path)
```

```python
for ids in subjids:
    dcm_folder = os.path.join(dicomfolder + ids)
    output_folder = os.path.join(niioutput + ids)
    cmd = os.path.join('dcm2niix -o ' + output_folder + ' -f %p ' + dcm_folder)
    print(cmd)
    if os.path.exists(output_folder):
        os.system(cmd)
    else: 
        os.mkdir(output_folder)
        os.system(cmd)
    
```

```python
def copyCertainFiles(source_folder, dest_folder, string_to_match, file_type=None):
    # Check all files in source_folder
    for filename in os.listdir(source_folder):
        # Move the file if the filename contains the string to match
        if file_type == None:
            if string_to_match in filename:
                shutil.move(os.path.join(source_folder, filename), dest_folder)

for ids in subjids:
    nii_folder = niioutput + ids
    anat = niioutput + ids + '/anat'  
    dwi = niioutput + ids + '/dwi' 
    fmap = niioutput + ids + '/fmap' 
    fmri = niioutput + ids + '/fmri' 
    others = niioutput + ids + '/others' 

    copyCertainFiles(nii_folder, anat, "t1" , file_type=None)      
    copyCertainFiles(nii_folder, anat, "t2" , file_type=None) 
    copyCertainFiles(nii_folder, others, "AAHead" , file_type=None) 
    copyCertainFiles(nii_folder, dwi, "DKI" , file_type=None) 
    copyCertainFiles(nii_folder, dwi, "FBI" , file_type=None)
    copyCertainFiles(nii_folder, dwi, "TDE" , file_type=None)
    copyCertainFiles(nii_folder, fmri, "sms3" , file_type=None)
    copyCertainFiles(nii_folder, fmap, "gre" , file_type=None)
```
